```python
import numpy as np
```

```python
arr1=np.random.randint(120,650,size=(4,7))
print(arr1)
```

```python
print(arr1.shape)
print(arr1.ndim)
print(arr1.size)
print(arr1.dtype)
```

Part-B

```python
avgperhos=np.mean(arr1,axis=1)
max_perhos=np.max(arr1,axis=1)
highest_avg=np.max(avgperhos)
print(avgperhos)
print(max_perhos)
print(highest_avg)
```

```python
total_per_day=np.sum(arr1,axis=0)
print(total_per_day)
highest_day=np.argmax(total_per_day)
print(highest_day)
```

Part-C


```python
penalized_usage=np.where(arr1>500,arr1*1.20,arr1)
print(penalized_usage)
```

Part-D

```python
hostel_avg_col=avgperhos[:,np.newaxis]
above_avg=arr1>hostel_avg_col
print(above_avg)
```

Part-E : Reshaping & Data Structure

```python
single_row=arr1.reshape(1,28)
print(single_row)
```

```python
two_rows=arr1.reshape(2,14)
print(two_rows)
```

Part-F : Memory Safety

```python
ref_backup=arr1
safe_backup=arr1.copy()
arr1[0,0]=999
print("Original array after modification:")
print(arr1)
print("Reference backup (affected by change):")
print(ref_backup)
print("Safe backup (unaffected by change):")
print(safe_backup)
```
