```python

mark=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(mark)
```

```python
print(mark.shape)
print(mark.size)
print(mark.dtype)
print(mark[1,:])
print(mark[:,2:])
print(mark[0][1])
```

Part-B

```python
from matplotlib.pylab import where


print(np.mean(mark,axis=0)) #To find the mean of columns
print(np.mean(mark,axis=1)) #To find the mean of rows
print(np.mean(mark[mark>=80])) #To find the mean of values greater than 80
print(mark+5)
```

Part-c Conditional Logic

```python
print(np.where(mark>=50,"1","0"))
print(np.where(mark<40,"0",mark))
```

Part-D

```python
av=np.mean(mark,axis=1)
print(av)
print(np.where(av>=85,"A",np.where(av>=70,"B",np.where(av>=50,"C","No Grade"))))
```

Part-E Data Integrity

```python
mark1=mark
mark2=mark.copy()
mark[0,0]=0
print(mark1)
print(mark2)
mark2[0,0]=100
print(mark1)
```
